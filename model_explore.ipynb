```python
import random
import torch
import torch.nn.functional as F
from torch_geometric.data import Data
from torch_geometric.datasets import Planetoid
from torch_geometric.nn import GCNConv
from torch_geometric.transforms import NormalizeFeatures
from torch_geometric.utils import to_undirected

class GCN(torch.nn.Module):
    def __init__(self, in_channels, hidden_channels, out_channels):
        super(GCN, self).__init__()
        self.conv1 = GCNConv(in_channels, hidden_channels)
        self.conv2 = GCNConv(hidden_channels, out_channels)

    def forward(self, x, edge_index):
        # x, edge_index = data.x, data.edge_index
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = F.dropout(x, training=self.training)
        x = self.conv2(x, edge_index)
        return F.log_softmax(x, dim=1)
```

```python
dataset = Planetoid(root='C:/data/CiteSeer', name='CiteSeer', transform=NormalizeFeatures())
print(len(dataset), '\n', dataset[0], '\n', dataset[0].is_directed())
print(dataset.num_node_features, dataset.num_classes)
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
data = dataset[0].to(device)
print(data)
print(data.x.shape, data.edge_index.shape)

gcn_model = GCN(dataset.num_node_features, 32, dataset.num_classes).to(device)
print(gcn_model)

print(dataset.num_features, dataset.num_classes)
print(data.x.shape)
print(data.edge_index)
print(dataset.y)
```
